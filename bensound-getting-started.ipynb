# Bensound API Demonstration

This is a very quick introduction to this API, which allows you to interact with the www.bensound.com website via Python. A few things you can do:
- download music
- download song art
- get song metadata
- use as a media source for a streaming media player (`Song.get_song_stream`)

There is a lot you can do. This is a very rough start. It will improve with time. So, make suggestions, and help the project along if you wish.

```python
from bensound import BensoundAPI
```

```python
# create an api object
api = BensoundAPI()
```

```python
# extract all data from website (not mp3 files)
api.extract_all_data()
```

```python
# see a list of available channels
api.get_channel_list()
```

```python
# see a list of available songs
api.get_song_list()
```

```python
# see songs available on a specific channel
api.get_songs_by_channel('Electronica')
```

```python
# get a specific song object
song = api.get_song_by_title('Epic Ride')
```

```python
# see the song properties
song.get_properties()
```

```python
# download the song to a destination or current working directory (default)
song.download_mp3()

# you can use the `song.get_song_stream` to create a `BytesIO` object that can be played 
# directly from a streaming media application that you build, such as with PyGame
```

```python
# get the image used in the song art
song.get_song_art()
```
